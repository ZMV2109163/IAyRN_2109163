```python
# Librerías a utilizar
import matplotlib.pyplot as plt # generar gráficos
import numpy as np # generar arreglos
import pandas as pd # cargar conjuntos de datos externos
import sklearn.linear_model # contiene diferentes modelos de ML
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
horas_estudio=np.abs(np.random.randn(100,1)*10)
calificaciones=2*horas_estudio+np.random.randn(100,1)*2
```

```python
X_train, X_test, y_train, y_test = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=100)
```

```python
model = sklearn.linear_model.LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrático medio: {mse:.2f}")
```

```python
plt.scatter(X_test, y_test, color='blue', label='Datos de prueba')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regresión lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Modelo de Regresión Lineal')
plt.legend()
plt.show()
```
